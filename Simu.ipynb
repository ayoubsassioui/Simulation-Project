```python
def client_label(type_client):
    if type_client<0.5:
        return 'A'
    else: 
        return 'B'
    
def label_temps_inter(d):
    if d<0.25:
        return 2
    elif d<0.5:
        return 3
    elif d<0.75:
        return 4
    else :
        return 5
  
    
def temps_arriv(d):
    l=[]
    l.append(d[0])
    for i in range(1,len(d)):
        l.append(d[i]+l[i-1])
    return l
def goto2(g):
    if g<0.5:
        return 0
    else :return 1
    
def goto1(g):
    if g<0.25:
        return 0
    else :return 1

def label_temps_service(d):
    if d<0.2:
        return 1
    elif d<0.4:
        return 2
    elif d<0.6:
        return 3
    elif d<0.8:
        return 4
    else :
        return 5

   

def debut_fin_service(temps_service,temps_arr):
    d=[]
    f=[]
    temps_service=temps_service.values.tolist()
    temps_arr=temps_arr.values.tolist()
    d.append(temps_arr[0])
    f.append(temps_arr[0]+temps_service[0])
    for i in range(1,len(temps_arr)):
        d.append(max(temps_arr[i],f[i-1]))
        x=d[i]+temps_service[i]
        f.append(x)
    return d,f


def temps_attent(d,a):
    l=[]
    d=d.values.tolist()
    a=a.values.tolist()
    for i in range(len(d)):
        l.append(max(0,d[i]-a[i]))
    return l  

def temps_system(d,f):
    l=[]
    d=d.values.tolist()
    f=f.values.tolist()
    for i in range(len(d)):
        l.append(f[i]-d[i])
    return l  

def serveur_libre(a,f):
    l=[]
    a=a.values.tolist()
    f=f.values.tolist()
    l.append(0)
    for i in range(1,len(a)):
        l.append(max(0,a[i]-f[i-1]))
    return l  
```
